# COVID DATA READER

## Importing required header files

```python
import numpy as np
import pandas as pd

# plotly packages
import plotly
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

init_notebook_mode(connected=True)
cf.go_offline()
```

## Getting the data from WHO website 

```python
data=pd.read_csv("https://covid19.who.int/WHO-COVID-19-global-data.csv")
```

## Making a function to take the country name and print the data

```python
def printdata(name):
    df = data[data['Country'] == name]
    df.iplot(x='Date_reported', y='New_cases')
```

```python
printdata("India")
```
